```python
import sys
sys.path.append('../src/')

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from DataSetFactory import DataSetFactory
from AMAP import AMAP
```

```python
RANDOM_STATE = 42
```

```python
ds = DataSetFactory.get_dataset('FashionMNIST64', random_state=RANDOM_STATE, is_subset=True, sample=0.15)
```

```python
X = ds.df[ds.feature_cols].values
y = ds.df[ds.label_col].values
```

# PCA

```python
amap = AMAP(dim_reduction_algo='pca',
            reduce_all_points=True,
            n_iter=1, 
            batch_size=0,
            learning_rate=0.1,
            do_animation=False,
            dataset='FashionMNIST',
            class_to_label=ds.class_to_label,
            show_fig=True,
            save_fig=True,
            random_state=RANDOM_STATE,
            show_anchors=False,
            supervised=False,
            show_points=True,
            show_inner_blobs=False,
            show_polygons=False,
            show_label_level_plots=False,
            show_anchor_level_plot=False,
            annotate_images=False)
print(amap)
```

```python
pca_low_dim = amap.fit_transform(X, y)
```

# UMAP

```python
umap_amap = AMAP(dim_reduction_algo='umap',
            reduce_all_points=True,
            n_iter=1, 
            batch_size=0,
            learning_rate=0.1,
            do_animation=False,
            dataset='FashionMNIST',
            class_to_label=ds.class_to_label,
            show_fig=True,
            save_fig=True,
            random_state=RANDOM_STATE,
            show_anchors=False,
            supervised=True,
            show_points=True,
            show_inner_blobs=False,
            show_polygons=False,
            show_label_level_plots=False,
            show_anchor_level_plot=False,
            annotate_images=False)
print(umap_amap)
```

```python
umap_low_dim = umap_amap.fit_transform(X, y)
```

# TSNE

```python
tsne_amap = AMAP(dim_reduction_algo='umap',
            reduce_all_points=True,
            n_iter=1, 
            batch_size=0,
            learning_rate=0.1,
            do_animation=False,
            dataset='FashionMNIST',
            class_to_label=ds.class_to_label,
            show_fig=True,
            save_fig=True,
            random_state=RANDOM_STATE,
            show_anchors=False,
            supervised=False,
            show_points=True,
            show_inner_blobs=False,
            show_polygons=False,
            show_label_level_plots=False,
            show_anchor_level_plot=False,
            annotate_images=False)
print(tsne_amap)
```

```python
tsne_low_dim = tsne_amap.fit_transform(X, y)
```

# Our Blobs No Relaxation

```python
blobs_amap = AMAP(dim_reduction_algo='umap',
                reduce_all_points=False,
                n_iter=1, 
                batch_size=0,
                learning_rate=0.1,
                do_animation=False,
                dataset='FashionMNIST',
                class_to_label=ds.class_to_label,
                show_fig=True,
                save_fig=True,
                random_state=RANDOM_STATE,
                show_anchors=False,
                supervised=False,
                show_points=False,
                show_inner_blobs=True,
                show_polygons=True,
                show_label_level_plots=False,
                show_anchor_level_plot=False,
                annotate_images=False,
                birch_threshold=0.6,
#                 birch_threshold=1,
#                 alpha=None,
                alpha=0.6,
                douglas_peucker_tolerance=0.3,
                smooth_iter=3,
                only_inter_relations=False,
                mask_relation_in_same_label=True,)
print(blobs_amap)
```

```python
blobs_low_dim = blobs_amap.fit_transform(X, y)
```

```python
blobs_amap.alpha=[0,1.3,0.8,1]
blobs_amap.douglas_peucker_tolerance = 0.3
blobs_amap.skip_polygons_with_area = 0.5
blobs_amap.anchors_plot_sns_separate(1, True)
```
