```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from DataSetFactory import DataSetFactory
from AMAP import AMAP
```

```python
RANDOM_STATE = 42
```

```python
ds = DataSetFactory.get_dataset('FashionMNIST64', random_state=RANDOM_STATE, sample=0.15, is_subset=True)
```

```python
X = ds.df[ds.feature_cols].values
y = ds.df[ds.label_col].values
```

```python
amap = AMAP(learning_rate=0.5, 
            n_iter=51, 
            batch_size=1,
            anchors_method='birch', 
            birch_threshold=0.6,
            umap_n_neighbors=5,
            dataset='FashionMNIST_64_15_perc',
            class_to_label=ds.class_to_label,
            radius_q=None,
            uniform_points_per='anchor',
            show_fig=True,
            save_fig=True,
            k=40,
            loss='Linf',
            random_state=RANDOM_STATE,
            do_animation=False,
            save_fig_every=50,
            magnitude_step=True,
            only_inter_relations=True,
            top_greedy=10,
            show_anchors=False,
            supervised=False,
            alpha=1,
            douglas_peucker_tolerance=0.5,
            smooth_iter=3,
            annotate_images=False,
            show_inner_blobs=True)
print(amap)
```

```python
low_dim = amap.fit_transform(X, y)
```
